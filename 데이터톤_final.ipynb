```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_age_by_metabolic(df):
    """
    대사증후군 판정(0/1)에 따른 나이 분포를 boxplot으로 시각화하는 함수
    - df: DataFrame (대사증후군_판정, 나이 컬럼 포함)
    """

    plt.figure(figsize=(8, 6))
    sns.boxplot(
        x=df['대사증후군_판정'],
        y=df['나이'],
        palette=['#B0B0B0', '#0052cc']  # 회색 / 파랑
    )

    plt.title("나이에 따른 대사증후군", fontsize=14)
    plt.ylabel("나이", fontsize=12)

    plt.xticks([0, 1], ['정상', '대사증후군'])
    plt.xlabel("")
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

# 사용 예시
plot_age_by_metabolic(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_uricacid_by_metabolic(df):
    """
    대사증후군 유무(0/1)에 따른 요산 수치 분포를 boxplot으로 시각화하는 함수
    - df: DataFrame (대사증후군_판정, 요산 컬럼 포함)
    """

    plt.figure(figsize=(8, 6))
    sns.boxplot(
        x=df['대사증후군_판정'],
        y=df['요산'],
        palette=['#B0B0B0', '#0052cc']  # 정상=회색, 대사증후군=파랑
    )

    plt.title("요산 수치에 따른 대사증후군", fontsize=14)
    plt.ylabel("요산", fontsize=12)

    plt.xticks([0, 1], ['정상', '대사증후군'])  # x축 레이블 변경
    plt.xlabel("")  # 원래 컬럼명 표시 제거
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

# 사용 예시
plot_uricacid_by_metabolic(df)
```
